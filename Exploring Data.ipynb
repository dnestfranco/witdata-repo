# Plots with Matplotlib

```python
# importing the most important libraries in data visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 9, 20) # creates a lit of 20 numbers from -10 to 9; similar to range()
print(x)
```

```python
y = x ** 2 # squaring all the numbers in x
print(y)
```

```python
plt.plot(x, y, 'o') # this will create an x-y graph
```

```python
plt.plot(x, y, 'b')

# adding labels
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Squared Function')
plt.show() # applying changes
```

```python
z = x ** 3 # cubing the numbers in x

plt.plot(x, z, 'b')

# adding labels
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Cubed Function')
plt.show() # applying changes
```

# Using Datasets to Create Barplots

```python
# importing the most important libraries in data visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

We created some data from student information in our period 1 class. We will now upload that data and visualize the information.

```python
data = pd.read_csv('data.csv')
```

```python
data.head(10)
```

Creating a histogram. A histogram is a chart of bars that shows us how our numeric data is distributed.

```python
sleep = data['Hours_of_Sleep'].values #this is getting the numeric values of our sleep data
print(sleep)

bins = np.arange(min(sleep), max(sleep) + 2)
print(bins)
```

```python
plt.hist(sleep, bins, histtype='bar', rwidth=0.9)

plt.title("Sleep Hours in Period 1")
plt.show()
```
